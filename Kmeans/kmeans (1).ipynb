```python
import numpy as np
import random

from sklearn.datasets import fetch_california_housing
```

# Load Dataset

```python
data = fetch_california_housing()
X = data.data
sample_numbers = 4000
indices = np.random.choice(X.shape[0], sample_numbers, replace=False)
x_data = X[indices]
```

# Implement KMeans algorithm (do not use any libraries and implement it from scratch)

```python
class KMeans:
    def __init__(self, n_clusters=5, max_iter=1000):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X):
        self.centroids = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]
        for _ in range(self.max_iter):
            self.labels = [self._closest_centroid(x) for x in X]
            new_centroids = np.array([X[np.array(self.labels) == i].mean(axis=0) for i in range(self.n_clusters)])
            if np.all(self.centroids == new_centroids):
                break
            self.centroids = new_centroids

    def _closest_centroid(self, x):
        return np.argmin([np.linalg.norm(x - centroid) for centroid in self.centroids])

    def predict(self, X):
        return [self._closest_centroid(x) for x in X]
```

# Use elbow method and silhouette method to choose optimal cluster numbers

```python
#TODO

import matplotlib.pyplot as plt

def elbow_method(X, max_clusters=10):
    inertia = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        inertia.append(np.sum([np.linalg.norm(x - kmeans.centroids[kmeans.labels[i]])**2 for i, x in enumerate(X)]))

    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_clusters + 1), inertia, 'bo-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method For Optimal k')
    plt.show()

elbow_method(x_data)
```

```python
from sklearn.metrics import silhouette_score

def silhouette_method(X, max_clusters=10):
    silhouette_avg = []
    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        cluster_labels = kmeans.predict(X)
        silhouette_avg.append(silhouette_score(X, cluster_labels))

    plt.figure(figsize=(8, 5))
    plt.plot(range(2, max_clusters + 1), silhouette_avg, 'bo-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Average silhouette score')
    plt.title('Silhouette Method For Optimal k')
    plt.show()

silhouette_method(x_data)
```

# Compare your k means algorithm with sklearn k means algorithm (compare it with both elbow and silhouette method)

```python
#TODO
from sklearn.cluster import KMeans as SKLearnKMeans

def compare_kmeans(X, max_clusters=10):
    # Elbow Method
    inertia_custom = []
    inertia_sklearn = []
    for k in range(1, max_clusters + 1):
        kmeans_custom = KMeans(n_clusters=k)
        kmeans_custom.fit(X)
        inertia_custom.append(np.sum([np.linalg.norm(x - kmeans_custom.centroids[kmeans_custom.labels[i]])**2 for i, x in enumerate(X)]))

        kmeans_sklearn = SKLearnKMeans(n_clusters=k)
        kmeans_sklearn.fit(X)
        inertia_sklearn.append(kmeans_sklearn.inertia_)

    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_clusters + 1), inertia_custom, 'bo-', label='Custom KMeans')
    plt.plot(range(1, max_clusters + 1), inertia_sklearn, 'ro-', label='Sklearn KMeans')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method Comparison')
    plt.legend()
    plt.show()

    # Silhouette Method
    silhouette_custom = []
    silhouette_sklearn = []
    for k in range(2, max_clusters + 1):
        kmeans_custom = KMeans(n_clusters=k)
        kmeans_custom.fit(X)
        cluster_labels_custom = kmeans_custom.predict(X)
        silhouette_custom.append(silhouette_score(X, cluster_labels_custom))

        kmeans_sklearn = SKLearnKMeans(n_clusters=k)
        kmeans_sklearn.fit(X)
        silhouette_sklearn.append(silhouette_score(X, kmeans_sklearn.labels_))

    plt.figure(figsize=(8, 5))
    plt.plot(range(2, max_clusters + 1), silhouette_custom, 'bo-', label='Custom KMeans')
    plt.plot(range(2, max_clusters + 1), silhouette_sklearn, 'ro-', label='Sklearn KMeans')
    plt.xlabel('Number of clusters')
    plt.ylabel('Average silhouette score')
    plt.title('Silhouette Method Comparison')
    plt.legend()
    plt.show()

compare_kmeans(x_data)
```

#### I prefer silhouette method because in elbow method sometimes it's hard to say where is the elbow point but silhouette method is more deterministic rather than elbow method

```python
# Step 1: Determine optimal cluster number using Elbow Method for Custom KMeans and Sklearn KMeans
def get_optimal_clusters_elbow(inertia):
    # This function helps to find the elbow point
    diff = np.diff(inertia)
    second_diff = np.diff(diff)
    return np.argmax(second_diff) + 2  # +2 because the second derivative array is 2 elements shorter

# Step 2: Determine optimal cluster number using Silhouette Method for Custom KMeans and Sklearn KMeans
def get_optimal_clusters_silhouette(silhouette_avg):
    return np.argmax(silhouette_avg) + 2  # +2 because the silhouette array starts from 2 clusters

# Using the previously defined comparison function to get inertia and silhouette scores
inertia_custom = []
inertia_sklearn = []
silhouette_custom = []
silhouette_sklearn = []

for k in range(1, 11):
    kmeans_custom = KMeans(n_clusters=k)
    kmeans_custom.fit(x_data)
    inertia_custom.append(np.sum([np.linalg.norm(x - kmeans_custom.centroids[kmeans_custom.labels[i]])**2 for i, x in enumerate(x_data)]))

    kmeans_sklearn = SKLearnKMeans(n_clusters=k)
    kmeans_sklearn.fit(x_data)
    inertia_sklearn.append(kmeans_sklearn.inertia_)

    if k >= 2:
        cluster_labels_custom = kmeans_custom.predict(x_data)
        silhouette_custom.append(silhouette_score(x_data, cluster_labels_custom))

        silhouette_sklearn.append(silhouette_score(x_data, kmeans_sklearn.labels_))

# Find optimal clusters
optimal_clusters_custom_elbow = get_optimal_clusters_elbow(inertia_custom)
optimal_clusters_sklearn_elbow = get_optimal_clusters_elbow(inertia_sklearn)
optimal_clusters_custom_silhouette = get_optimal_clusters_silhouette(silhouette_custom)
optimal_clusters_sklearn_silhouette = get_optimal_clusters_silhouette(silhouette_sklearn)

print(f"Optimal clusters based on elbow method for Custom KMeans: {optimal_clusters_custom_elbow}")
print(f"Optimal clusters based on elbow method for Sklearn KMeans: {optimal_clusters_sklearn_elbow}")
print(f"Optimal clusters based on silhouette method for Custom KMeans: {optimal_clusters_custom_silhouette}")
print(f"Optimal clusters based on silhouette method for Sklearn KMeans: {optimal_clusters_sklearn_silhouette}")

# Comparison and explanation
if optimal_clusters_custom_elbow != optimal_clusters_sklearn_elbow:
    print(f"Different optimal cluster numbers based on elbow method: Custom KMeans = {optimal_clusters_custom_elbow}, Sklearn KMeans = {optimal_clusters_sklearn_elbow}")

if optimal_clusters_custom_silhouette != optimal_clusters_sklearn_silhouette:
    print(f"Different optimal cluster numbers based on silhouette method: Custom KMeans = {optimal_clusters_custom_silhouette}, Sklearn KMeans = {optimal_clusters_sklearn_silhouette}")
```
